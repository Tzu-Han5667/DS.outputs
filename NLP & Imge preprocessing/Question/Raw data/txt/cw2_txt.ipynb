```python
#book content handling
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()
import csv
book_list=['puck-of-pooks-hill.txt','man-who-would-be-king.txt','kim.txt','just-so-stories.txt',\
           'jungle-book.txt','ginger-pickles.txt','jeremy-fisher.txt','squirrel-nutkin.txt',\
           'benjamin-bunny.txt','peter-rabbit.txt']

bk_list={}

for i in book_list:
    b_k=[]
    with open(i) as book:
        book=csv.reader(book)
        bk=list(book)
        for j in bk:
            for k in j:
                word=tknzr.tokenize(k.strip())
                for l in word:
                    b_k.append(l)
    bk_list[i]=b_k
```

```python
#stopwords
import csv
import numpy as np
character=list('abcdefghijklmnopqrstuvwxyz')
chars=[]
for i in character:
    chars.append(list(i))
data=[]
with open('english-stop-words-large.txt') as file:
    files=csv.reader(file)
    data=list(files)
content=[]
for i in data:
    for j in i:
        content.append(j)
for i in character:
    content.remove(i)
error=['',"\n;\n<NUM>\n?\n$\nkm\ns\nu\n&\n#\n's\n/\ndr.\n"]
num=[',','?','$','km','s','u','&','#',"'s",'/','dr.']
stopword=list(set(content)-set(error))
for i in num:
    stopword.append(i)
stopword=np.array(stopword)
```
